```python
import os
import numpy as np
import keras
from keras import layers, Sequential
from tensorflow import data as tf_data
import tensorflow as tf
import matplotlib.pyplot as plt




image_size = (300, 300)
batch_size = 128

train_ds, val_ds = keras.utils.image_dataset_from_directory(
    "Images",
    validation_split=0.2,
    subset="both",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)

classCount = len(train_ds.class_names)

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(np.array(images[i]).astype("uint8"))
        plt.title(int(labels[i]))
        plt.axis("off")



inputShape = (300, 300, 3)


def makeModel(inputShape, classCount) :
    input = layers.Input(shape=inputShape)
    x = layers.Rescaling(1./255)(input)
    x=layers.Conv2D(5, 3, strides=3, padding='same', activation='relu')(x)
    x=layers.Conv2D(15, 3, strides = 2, padding='same')(x)
    x=layers.Conv2D(35, 3, strides = 1, padding='same', activation='relu')(x)
    x=layers.Conv2D(85, 3, strides = 1, padding='same')(x)
    x=layers.Conv2D(185, 5, strides = 1, padding='same')(x)
    x=layers.MaxPooling2D()(x)
    x = layers.Dropout(0.25)(x)
    x=layers.Flatten()(x)
    outputs = layers.Dense(classCount)(x)
    return keras.Model(input, outputs)



model = makeModel(inputShape, classCount)

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
epochs=35

model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
model.save('model.keras', overwrite=True)
print("Model saved")
```
